1. What is a random variable in probability theory?

A random variable is a variable that represents the numerical outcome of a random experiment. Its value depends on chance. For example, when a coin is tossed, the outcome (0 for tail and 1 for head) can be represented using a random variable.

2. What are the types of random variables?

There are two main types of random variables:

Discrete random variables, which take countable values.

Continuous random variables, which take values from a continuous range.

3. What is the difference between discrete and continuous distributions?

A discrete distribution deals with countable outcomes such as number of students or number of heads in coin tosses. A continuous distribution deals with measurable values like height, weight, or time, where infinite values are possible within a range.

4. What are probability distribution functions (PDF)?

A probability distribution function describes how probabilities are assigned to different values of a random variable. It shows the likelihood of occurrence of each possible value and helps in understanding data behavior.

5. How does the cumulative distribution function (CDF) differ from the PDF?

A PDF gives the probability density at a specific value, while a CDF gives the probability that a random variable is less than or equal to a certain value. The CDF is obtained by integrating the PDF.

6. What is a discrete uniform distribution?

A discrete uniform distribution is a distribution in which all possible outcomes have equal probability. An example is rolling a fair die where each face has the same chance of occurring.

7. What are the key properties of a Bernoulli distribution?

A Bernoulli distribution has only two possible outcomes: success and failure. The probability of success is denoted by p, and failure by (1 − p). It is used in experiments like a single coin toss.

8. What is the binomial distribution and how is it used?

The binomial distribution represents the number of successes in a fixed number of independent trials. Each trial has only two outcomes. It is widely used in quality testing and probability experiments.

9. What is the Poisson distribution and where is it applied?

The Poisson distribution measures the number of events occurring in a fixed interval of time or space. It is commonly used in traffic analysis, call centers, and arrival problems.

10. What is a continuous uniform distribution?

A continuous uniform distribution is a distribution where all values in a given range have equal probability. It is used when outcomes are equally likely within an interval, such as random numbers between 0 and 1.

11. What are the characteristics of a normal distribution?

A normal distribution is a symmetric, bell-shaped curve. In this distribution, the mean, median, and mode are equal. It is completely described by its mean and standard deviation.

12. What is the standard normal distribution and why is it important?

The standard normal distribution is a normal distribution with a mean of 0 and a standard deviation of 1. It is important because it allows comparison of different datasets using Z-scores.




13. What is the Central Limit Theorem (CLT)?

The Central Limit Theorem states that the distribution of sample means tends to follow a normal distribution as the sample size increases, regardless of the original population distribution.

14. How does the Central Limit Theorem relate to the normal distribution?

The CLT explains why many natural and real-world datasets follow a normal distribution. It forms the foundation for many statistical methods and tests.

15. What is the application of Z-statistics in hypothesis testing?

Z-statistics are used to determine whether to accept or reject a null hypothesis. They are mainly applied when sample size is large and population variance is known.

16. How do you calculate a Z-score and what does it represent?

A Z-score is calculated using the formula
Z = (X − μ) / σ
It represents how far a data point is from the mean in terms of standard deviation.

17. What are point estimates and interval estimates in statistics?

A point estimate provides a single value as an estimate of a population parameter, while an interval estimate provides a range of values within which the parameter is likely to lie.

18. What is the significance of confidence intervals in statistical analysis?

Confidence intervals show the reliability of an estimate and indicate the range in which the true population parameter is expected to exist with a certain level of confidence.

19. What is the relationship between a Z-score and a confidence interval?

Z-scores are used to calculate the confidence interval. Higher confidence levels require larger Z-values, resulting in wider intervals.

20. How are Z-scores used to compare different distributions?

Z-scores standardize data, making it possible to compare values from different distributions with different means and standard deviations.

21. What are the assumptions for applying the Central Limit Theorem?

The CLT assumes random sampling, independence of observations, and a sufficiently large sample size for accurate results.

22. What is the concept of expected value in probability?

Expected value is the average outcome of a random variable over a large number of trials. It represents the long-term expected result.

23. How does a probability distribution relate to the expected outcome of a random variable?

A probability distribution assigns probabilities to outcomes, and these probabilities are used to calculate the expected value, which summarizes the overall behavior of the random variable.
